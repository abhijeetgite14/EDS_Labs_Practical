# KN mean and KNNmean

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/esheet.xlsx')
print("name of emp")
x = df["name of emp"]
print(x)
x1 = list(x)
print(x1)
print()
print("emp id")
y = df["emp id"]
print(y)
y1 = list(y)
plt.scatter(x1,y1)
plt.show()
data = list(zip(x1,y1))
inertia = []
for i in range(1,120):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(data)
  inertia.append(kmeans.inertia_)
plt.plot(range(1,120), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
kmeans = KMeans(n_clusters=5)
kmeans.fit(data)
plt.scatter(x1,y1,c = kmeans.labels_)
plt.show()
```
